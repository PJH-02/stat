```python
import numpy as np
from matplotlib import pyplot as plt

rnd1=200+np.random.randn(100)
x=[x for x in range(len(rnd1))]

plt.plot(x, rnd1, '-')
plt.fill_between(x, rnd1, 195, where=(rnd1>195), facecolor='g', alpha=0.6)

plt.title("Example1")
plt.show()
```
